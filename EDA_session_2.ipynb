**create data frame using dictionary**

```python
name=['Avinash','akash','aadhya']
age=[32,28,31]
pd.DataFrame(zip(name,age),columns=["names","ages"])
```

```python
dict1={'Names':['Avinash','akash','aadhya'],"ages":[32,28,31]}
pd.DataFrame(zip(dict1['Names'],dict1['ages']),columns=["names","ages"])
```

```python
pd.DataFrame(dict1)
```

```python
dict2={'Names':'Avinash',"ages":32}
```

```python
pd.DataFrame(dict2,index=[1])
```

```python
pd.DataFrame(dict2,index=[0])
```
