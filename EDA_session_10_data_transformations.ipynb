**Data transformations**

- Generally data has 3 types of distribution

  - normal distibution

  - pos skewed

  - neg skewed

- All math concepts are developed by assuming as data follows Normal distribution

- Data skew happens because of many reasons,in that one reason is outliers

- Once we perform the outlier analysis then we can apply transformation methods,if the data does not follows normality

- We have some transformation methods mentioned below
  
  - log transformation
 
  - exponential trans
 
  - reciprocal trans
 
  - square root trans
 
  - power trans

**Log transformation**

- Log transformation means performing logarithm operations on origianl data

- It is an approach to convert into Noraml distribution

- Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
```

```python
plt.hist(log_data,bins=20,label="Exponential")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=30,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=30,label='Logaritham')
plt.legend() #it gives the scale inside the barchart
plt.show()
```

**Reciprocal Transformation**

- Reciprocal will fail when values are low

- For examples x=0 then 1/0 fail

```python
x=2
1/x
```

```python
resiprocal_data=np.reciprocal(exp_data)
plt.hist(resiprocal_data,bins=20,label="Reciprocal")
plt.show()
```

**Square root transformation**

```python
square_root_data = np.sqrt(exp_data)
plt.hist(square_root_data, bins=20, label="Square Root Data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
log_data = np.log(exp_data)
plt.subplot(1,3,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
log_data = np.log(exp_data)
plt.subplot(1,3,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
square_data = np.sqrt(exp_data)
plt.subplot(1,3,3).hist(square_data,bins=20,label='Square')
plt.legend()
plt.show()
```

**Power Transformation**

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
# save
# apply fit transform
```

```python
from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans = power.fit_transform(exp_data1)
plt.hist(data_trans)
```
