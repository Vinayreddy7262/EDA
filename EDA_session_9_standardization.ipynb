```python
#Task
#step-1:take previling wage column 

#step-2:Calculate the mean value of that column

#step-3:calculate the standart deviation of the previlling wage

#step-4:calculate the Nr:step1-step2

#step-5:divide the step4/step3
```

```python
a=visa_df['prevailing_wage']
mean=visa_df['prevailing_wage'].mean()
std=visa_df['prevailing_wage'].std()
b=(a-mean)/std
visa_df['prevailing_wage_z']=b
visa_df
```
