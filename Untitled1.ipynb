```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv("Visadataset.csv")
```

**Map Method**

- first get the unique labels for each columns

- create a dictionary by providing a values to each label

- for example case status has two labels ,so we are providing 2 values

- Assign 0 for certified

- Assign 1 for Deniend

```python
cat=visa_df.select_dtypes(include='object').columns
cat
```

```python
visa_df['case_status'].unique()
```

```python
d={'Denied':1,'Certified':0}
d
```

```python
a=visa_df["case_status"].map(d)
```

```python
visa_df["case_status"]=a
```

```python
visa_df
```

```python
d1={'Y':1,'N':0}
```

```python
b=visa_df['has_job_experience'].map(d1)
```

```python
visa_df['has_job_experience']=b
visa_df
```

```python
for i in cat:
    labels=list(visa_df[i].unique())
    values=[i for i in range(len(labels))]
    d=dict(zip(labels,values))
    visa_df[i]=visa_df[i].map(d)

visa_df
```

**Label Encoder**

- label encoder is a package under **sickit learn**

- sickit-learn is heart of ML

- in side sickit-learn we have a method called **Preprocessing**

- under preprocessing we have **LableEncoder**

- scikit-learn we will write as **sklearn**

- any sklearn package we have 3 packages

- step-1: Read the package

- step-2: save the package

- step-3: Apply fit transform

```python
visa_df=pd.read_csv("Visadataset.csv")
```

```python
#step-1: Read the package

from sklearn.preprocessing import LabelEncoder

#step-2:save the package

le=LabelEncoder()

visa_df['case_status']=le.fit_transform(visa_df['case_status'])

```

```python
visa_df
```

```python
for i in cat:
    visa_df[i]=le.fit_transform(visa_df[i])

visa_df
```

```python
le.inverse_transform(visa_df['continent'])
```
